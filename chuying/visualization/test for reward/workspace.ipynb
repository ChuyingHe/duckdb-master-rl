```python
import os 
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import math
import pylab
import seaborn as sns
```

```python
cwd = os.getcwd()
csv = cwd + '/execution_time_0_to_1.csv'
data = pd.read_csv(csv)
data.head()
```

```python
print(data['Query'].unique())
```

```python
def distributionOfAttr(sql, attr):
    print(attr, max(data[data["Query"]==sql][attr]), min(data[data["Query"]==sql][attr]))
    sns.displot(data[data["Query"]==sql][attr])
    plt.savefig(attr+"-"+sql+".pdf")
```

```python
def showSQL(sql):
    distributionOfAttr(sql, "Time")
    # since the shorter the "Time" is, the better the reward should be
    # invert "Time" use (1/x)
    distributionOfAttr(sql, "Intermediate")
    distributionOfAttr(sql, "Reward")
```

```python
for sql in data['Query'].unique():
    showSQL(sql)
```
